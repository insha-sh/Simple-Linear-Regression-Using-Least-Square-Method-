# Linear Regression from Scratch

## Using least squares method

Formula : y = m * x + c

where 

y - response variable ;

x - explanatory variable ;

m - slope of the line ;

c - y-intercept ;


m = cov(x,y) / var(x)

c = mean(y) - m * mean(x)

```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Pre-processing

data = pd.read_csv("tvmarketing.csv")
X = data.iloc[: , 0]
Y = data.iloc[: , 1]
```

```python
# Plot the data 
plt.figure(figsize = [10,10])

plt.scatter(X,Y)
plt.title('TV vs Sales')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

```python
# Model building

X_mean = np.mean(X)
Y_mean = np.mean(Y)
N = len(X)

#Finding the Statistical Formula to calculate slope 'm' and y-intercept 'c'
m_num = 0
m_den = 0
for i in range(N):
    m_num += (X[i] - X_mean) * (Y[i] - Y_mean)
    m_den += (X[i] - X_mean)**2

m = m_num/m_den

c = Y_mean - (m * X_mean)

print(m, c)
```

```python
# Prediction the response variable

Y_pred = m*X + c

plt.figure(figsize = [10,10])
plt.scatter(X, Y, label = 'Scatter Plot')
plt.plot(X, Y_pred, color='red', label = "Regression Line")
plt.legend()
plt.show()
```

## Evaluating the model

1. Root mean square error

2. R-squared value

3. Residual Plot

```python
#1. Calculating Root Mean Squares Error

rmse = 0
for i in range(N):
    rmse += (Y[i] - Y_pred[i]) ** 2
rmse = np.sqrt(rmse/N)
print("RMSE")
print(rmse)
```

```python
#2. Calculating R2 Score
ss_tot = 0
ss_res = 0
for i in range(N):
    ss_tot += (Y[i] - Y_mean) ** 2
    ss_res += (Y[i] - Y_pred[i]) ** 2
r2 = 1 - (ss_res/ss_tot)
print("R2 Score")
print(r2)
```

```python
#3. Residual Plot

residuals = Y - Y_pred
res_y = [0] * N

plt.figure(figsize = [10,10])
plt.scatter(X, residuals, label = 'Scatter Plot')
plt.plot(X, res_y, color = 'red')
plt.show()
```
